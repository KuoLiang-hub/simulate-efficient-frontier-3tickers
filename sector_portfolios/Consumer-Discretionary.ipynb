```python
# shows sector performance
# see details (https://www.marketbeat.com/stocks/sectors/consumer-discretionary/)
import pandas as pd
import pandas_datareader.data as web
from matplotlib import pyplot as plt
```

```python
# using yahoo finance data
# S&P 500 Consumer Discretionary (Sector) (^SP500-25)

# a dictionary of sector and ticker
sector_dictionary = {'^SP500-25':'Consumer Discretionary'}

sectors = ['^SP500-25','EBAY','FL','BBY','SBUX','RH']
sector_data = pd.DataFrame()

for a in sectors:
    #you can chagne the start date to any date in 'yyyy-mm-dd' format.
    sector_data[a] = web.DataReader(a, data_source = 'yahoo', start = '2021-1-1')['Adj Close'] 

sector_data.rename(columns = sector_dictionary, inplace =True)
sector_data.head() #check the beginning of the data
```

```python
sector_data.tail() #check the end of the data
```

```python
(sector_data / sector_data.iloc[0]).plot(figsize=(12, 6)) #Plot of company performance without considering dividends
```

```python
(sector_data.iloc[-1] - sector_data.iloc[0]) / sector_data.iloc[0] #shows returns since begining date. 
```

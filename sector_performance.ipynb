```python
# using yahoo finance data
# S&P 500 Industrials (Sector) (^SP500-20)
# S&P 500 Information Technology (Sector) (^SP500-45)
# S&P 500 Health Care (Sector) (^SP500-35)
# S&P 500 Financials (Sector) (^SP500-40)
# S&P 500 Real Estate (Sector) (^SP500-60)
# S&P 500 Energy (Sector) (^GSPE)
# S&P 500 Materials (Sector) (^SP500-15)
# S&P 500 Consumer Discretionary (Sector) (^SP500-25)
# S&P 500 Utilities (Sector) (^SP500-55)
# S&P 500 Consumer Staples (Sector) (^SP500-30)
# S&P 500 Telecommunication Services (Sector) (^SP500-50)

# a dictionary of sector and ticker
sector_dictionary = {
    '^SP500-45' : 'Technology' ,
    '^SP500-20':'Industrials',
    '^SP500-35':'Health Care',
    '^SP500-40':'Financials',
    '^SP500-60':'Real Estate',
    '^GSPE':'Energy',
    '^SP500-15':'Materials',
    '^SP500-25':'Consumer Discretionary',
    '^SP500-55':'Utilities',
    '^SP500-30':'Consumer Staples',
    '^SP500-50':'Telecommunication Services'
    }

sectors = ['^SP500-45','^SP500-20','^SP500-35','^SP500-40','^SP500-60','^GSPE','^SP500-15','^SP500-25','^SP500-55','^SP500-30','^SP500-50']
sector_data = pd.DataFrame()

for a in sectors:
    #you can chagne the start date to any date in 'yyyy-mm-dd' format.
    sector_data[a] = web.DataReader(a, data_source = 'yahoo', start = '2020-1-1')['Adj Close'] 

sector_data.head() #check the beginning of the data
```

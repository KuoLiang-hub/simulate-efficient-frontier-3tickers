```python
# Download mutual fund index prices for the Thrift Savings Plan (TSP).
 
import pandas_datareader.tsp as tsp
```

```python
tspreader = tsp.TSPReader(start='2020-10-1', end='2022-4-30') #change start and end date here.
tspp = tspreader.read().sort_index()
```

```python
(tspp/tspp.iloc[0]).plot(figsize=(12,5))
```
